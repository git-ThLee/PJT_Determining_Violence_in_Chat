```python
import numpy as np
import pandas as pd
```

나눠진 csv파일을 한번에 불러왔습니다.

```python
df1 = '/content/drive/MyDrive/nlp pj/talksets_train_1.csv'
df2 = '/content/drive/MyDrive/nlp pj/talksets_train_2.csv'
df3 = '/content/drive/MyDrive/nlp pj/talksets_train_3.csv'
df4 = '/content/drive/MyDrive/nlp pj/talksets_train_4.csv'
df5 = '/content/drive/MyDrive/nlp pj/talksets_train_5.csv'
```

```python
df = pd.concat(
   map(pd.read_csv, [df1, df2, df3, df4, df5]), ignore_index=True)
print(df)
```

스마일 게이트 데이터와 합치기 위해 컬럼명을 통일했습니다.

```python
df['문장'] = df['text']
```

필요없는 컬럼을 삭제하고, 새로 컬럼을 생성합니다.

```python
df.drop('text',axis=1, inplace=True)
df.drop('types',axis=1, inplace=True)
df.drop('intensity',axis=1, inplace=True)
```

```python
df['clean'] = 0
```

```python
df['violence'] = 0
```

```python
df
```

True와 False 를 숫자 데이터로 변경해줍니다.

스마일게이트의 데이터 셋에 맞추어 float타입으로 변경합니다.

```python
df['is_immoral'] = df['is_immoral'].astype(float)
```

```python
df['violence'] = df['is_immoral']
```

```python
df
```

violence값이 0인 데이터는 clean한 데이터이기에 clean 값에 1을 넣어줍니다.

```python
for i in df[(df['violence'] == 0)].index:
        df.at[i, 'clean'] = 1
```

```python
df
```

```python
df['clean'] = df['clean'].astype(float)
```

```python
df
```
