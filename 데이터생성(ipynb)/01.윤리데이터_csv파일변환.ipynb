```python
import pandas as pd
import json
```

# json 파일 불러오기

```python
file_path = "/content/drive/MyDrive/NLP미니플젝-10조/Data/147.텍스트 윤리검증 데이터/01.데이터/2.Validation/라벨링데이터/aihub/talksets-train-6.json"

with open(file_path, 'r') as file:
    data = json.load(file)
    print(type(data))
    print(data)
```

# json to df 

```python
df = pd.DataFrame()

for i in range(len(data)):
  df = df.append(pd.json_normalize(data[i]['sentences']))
```

# 불필요 컬럼 제거

```python
df.drop('id',axis=1, inplace=True)
df.drop('speaker',axis=1, inplace=True)
df.drop('origin_text',axis=1, inplace=True)
df.drop('intensity_sum',axis=1, inplace=True)
df.drop('votes',axis=1, inplace=True)
df.drop('frame_id',axis=1, inplace=True)
df.drop('mapped_slots',axis=1, inplace=True)
```

# csv 파일 저장

```python
df.to_csv("talksets_valid.csv", index=False, encoding="utf-8-sig")
```
