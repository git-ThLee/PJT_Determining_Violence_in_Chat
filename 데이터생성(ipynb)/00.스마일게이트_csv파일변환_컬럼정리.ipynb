폭력적 데이터 고르기

```python
# 폭력적인 내용일 경우 '여성/가족', '남성', '성소수자', '인종/국적', '연령', '지역', '종교', '기타 혐오', '악플/욕설'의 값의 합이 1이상일 것이다
# 개인지칭의 경우 문제가 될 소지가 있기때문에 함께 넣었습니다.
```

```python
df_v = df[['여성/가족', '남성', '성소수자', '인종/국적', '연령', '지역', '종교', '기타 혐오', '악플/욕설', '개인지칭']]
```

```python
df_v['total'] = df_v.sum(axis=1)
```

```python
df_v
```

astype을 통해 0과 1로 데이터의 값을 변경했습니다.

```python
df_v['total'] = df_v['total'].astype(bool).astype(float)
```

```python
df_v
```

```python
df['violence'] = df_v['total']
```

```python
df
```

필요없는 컬럼 삭제

```python
df.drop('개인지칭',axis=1, inplace=True)
df.drop('악플/욕설',axis=1, inplace=True)
df.drop('기타 혐오',axis=1, inplace=True)
df.drop('종교',axis=1, inplace=True)
df.drop('지역',axis=1, inplace=True)
df.drop('연령',axis=1, inplace=True)
df.drop('인종/국적',axis=1, inplace=True)
df.drop('성소수자',axis=1, inplace=True)
df.drop('남성',axis=1, inplace=True)
df.drop('여성/가족',axis=1, inplace=True)
```
