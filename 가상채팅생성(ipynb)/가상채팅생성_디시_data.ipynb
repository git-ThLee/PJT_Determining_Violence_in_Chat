# 이상한 키워드 제거

```python
import re

dc_data2 = []

remove_words = [
    "\\", "르블랑", "리산","리신","야스오","힙붕","힙스터","히오스","희철","흥민","흡낫","강등","휴가",
    "정글", "정균","미드","서폿","탑","정보","원딜","정병","티어","클템", "힛갤","히오쑤왕","제이스","제드",
    "제라드","제라스","점화","그브","빵테","점멸","탈진"

]

# 특수문자 제거 & 길이 100 이상 제거(도배글)
for i in range(len(dc_data)):
  re_text = re.sub('[-=+,#/\?:^.@*\"※~ㆍ!』‘|\(\)\[\]`\'…》\”\“\’·]', ' ', dc_data[i])
  join_text = ' '.join(re_text.split())
  if len(join_text) < 101 :
    for word in remove_words:
      if join_text.find(word) == -1:
        dc_data2.append(join_text)
        #print(i, join_text)
        break

#dc_data2
```

# 파일 저장

```python
import random
```

```python
f = open("output.txt", 'w')
```

```python
users = ['짱구','철수',"유리",'맹구','훈이']
day = "오후 10:30"
message = ""
for _ in range(10000):
  random_message = '[' + random.choice(users) + '] ' + '[' + day + '] ' + random.choice(dc_data2)
  message = message + "\r\n"+ random_message
f.write("철수, 유리, 맹구 님과 카카오톡 대화\n저장한 날짜 : 2022-08-04 01:01:01\r\n\n")
f.write("--------------- 2022년 8월 4일 목요일 ---------------\r\n")
f.write("철수님이 유리님, 맹구님, 훈이님, 맹구님, 짱구님을 초대하였습니다.\r\n")
f.write(message)
```
